```python
# Importing Libraries #
import numpy as np
import pandas as pd
import os, csv, sys
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
```

```python
# Creating two random normal variables with different mean and std
X = 3*np.random.randn(10000) + 20 # This is N(20,3^2)
Y = 4*np.random.randn(10000) +4   # This is N(4, 4^2)
```

```python
# This variable is composed of a mixture of X & Y 
Z = np.array([X[i] if np.random.random() <0.6 else Y[i] for i in range(len(X))])
```

```python
print (np.mean(X), np.std(X), np.mean(Y), np.std(Y), np.mean(X+Y), np.std(X+Y), np.mean(Z), np.std(Z))
```

```python
plt.rcParams["figure.figsize"] = (15,8)
figure, axis = plt.subplots(2, 2)
axis[0,0].hist(X, bins = 50, density = True, color = 'Green')
axis[0,0].set_title('PDF of X')
axis[1,0].hist(Y, bins = 50, density = True, color ='red')
axis[1,0].set_title('PDF of Y')
axis[0,1].hist(X+Y, bins = 50, density = True, color = 'yellow')
axis[0,1].set_title('PDF of X+Y')
axis[1,1].hist(Z, bins = 50, density = True)
axis[1,1].set_title('PDF of Z (mixture of X&Y)')
#plt.hist(X, bins = 50)
plt.show()
```

```python
gm = GaussianMixture(n_components=2, random_state=0).fit(Z.reshape(-1,1))
```

```python
print ("Means: ",gm.means_)
print ("Weights: ", gm.weights_)
print ("Variance: ",gm.covariances_)
```
